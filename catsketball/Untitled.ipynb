```python
import stat_analysis
import pyarrow as pa
import pyarrow.compute as pc
```

```python
table = stat_analysis.load_projections()
```

```python
stdzers = stat_analysis.init_standardizers()
first_2_rounds = table.filter(pc.field("R#") < 29)
stat_analysis.fit_standardizers(first_2_rounds, stdzers)
first_2_rounds_stdzd = stat_analysis.standardize(first_2_rounds, stdzers)
(
    first_2_rounds_stdzd
    .drop_columns(["GP", "drafted_by", "MPG", "POS", "TEAM", *stat_analysis.POSITIONS])
    .sort_by("R#")
    .to_pandas()
)
```

```python
stdzers = stat_analysis.init_standardizers()
second_2_rounds = table.filter(
    (pc.field("R#") > 29)
    & (pc.field("R#") <= 56)
)
stat_analysis.fit_standardizers(second_2_rounds, stdzers)
second_2_rounds_stdzd = stat_analysis.standardize(second_2_rounds, stdzers)
(
    second_2_rounds_stdzd
    .drop_columns(["GP", "drafted_by", "MPG", "POS", "TEAM", *stat_analysis.POSITIONS])
    .sort_by("R#")
    .to_pandas()
)
```
